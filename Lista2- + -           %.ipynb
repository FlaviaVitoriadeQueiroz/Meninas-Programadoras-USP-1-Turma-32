```python
operação= input()
valor1= int (input())
valor2= int (input())
if operação== '+':
  calculo= valor1+valor2
  print(calculo)
elif operação== '-':
  calculo= valor1-valor2
  print(calculo)
elif operação== '*':
  calculo= valor1*valor2
  print(calculo)
elif operação== '/':
  calculo= valor1/valor2
  print(calculo)
elif operação== '%':
  calculo= valor1%valor2
  print(calculo)
elif operação== '**':
  calculo= valor1**valor2
  print(calculo)
elif operação== '//':
  calculo= valor1//valor2
  print(calculo)
```
