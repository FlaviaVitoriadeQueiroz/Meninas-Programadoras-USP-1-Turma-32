```python
massa = float(input())
aceleracao = float(input())
maior_aceleracao = aceleracao
voltas = 0
for _ in range(1000):
    forca = float(input())
    nova_aceleracao = forca / massa
    if nova_aceleracao < aceleracao:
        break
    aceleracao = nova_aceleracao
    if maior_aceleracao < aceleracao:
        maior_aceleracao = aceleracao
    voltas += 1
print(f"maior aceleração: {maior_aceleracao:.3f}")
print(f"número de voltas: {voltas}")
```
