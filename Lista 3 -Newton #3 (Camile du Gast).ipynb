```python
massa = float(input())
num_voltas = int(input())

maior_aceleracao = 0
volta_mais_rapida = 0

for i in range(1, num_voltas + 1):
    forca = float(input())
    aceleracao = forca / massa

    if aceleracao > maior_aceleracao:
        maior_aceleracao = aceleracao
        volta_mais_rapida = i

print('melhor_volta:', volta_mais_rapida)
print(f'maior aceleração:', '%.1f' % maior_aceleracao)
```
