```python
energy = input()
a = int(energy[0])
b = int(energy[-1])
if a == b:
  print('True')
else:
  print('False')
```

```python
energy = input()
soma = 0
for i in energy:
  soma += int(i)
print(soma)
```

```python
valora = int (input())
while soma < valora:
  tenta = int(input())
  soma += tenta
print(soma)
```
