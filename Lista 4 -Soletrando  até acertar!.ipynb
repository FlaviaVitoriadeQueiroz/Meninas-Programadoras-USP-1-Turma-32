```python
palavra = ''
certo = input()
acertou = 0

while acertou == 0:
  palavra = ''
  entrada = input()
  while entrada != '.':
    palavra = palavra + entrada
    entrada = input()
  if palavra == certo:
    print('8-)')
    acertou = 1
  else:
    print('8-|')
    acertou = 0
```
