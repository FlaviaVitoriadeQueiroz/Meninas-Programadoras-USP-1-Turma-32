```python
numero = int (input())
if numero == 0:
  print('.')
else:
  simbulo = '*'
  print(simbulo * numero)
```

```python
numero = input()
simbulo = '*'
for digito in numero:
  if '0' <= digito <= '9':
    print(simbulo * int(digito))
```
