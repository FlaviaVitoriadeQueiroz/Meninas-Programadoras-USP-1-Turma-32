```python
nvotos = int (input())
contf = 0
contc = 0
conto = 0
for i in range(nvotos):
  votos = input()
  if votos == 'F' or votos == 'f':
    contf += 1
  elif votos == 'C' or votos == 'c':
    contc += 1
  else:
    conto +=1
print('a favor:', contf)
print('contra:',contc)
print('outros:',conto)
```
