```python
senha = input() #entrada
cont = 0

for i in senha: #contando os caracteres
  cont = cont + 1
  tot = 0 #vendo se tem mais de 8 caracteres
  if cont >= 8:
     tot = 1

dig = "1234567890" #contando os digitos
contd = 0
for i in senha:
    if i in dig:
        contd = contd + 1
    tod = 0 #vendo se tem pelo menos dois digitos
    if contd >= 2:
      tod = 1

vogais = "aeiouAEIOU" #contando as v
contv = 0
for i in senha:
    if i in vogais:
        contv = contv + 1
    tov = 0
    if contv > contd: #vendo se tem mais v que digitos
       tov = 1

if tov + tot + tod == 3:
    print('OK')
else:
    print('ERRO')
```
