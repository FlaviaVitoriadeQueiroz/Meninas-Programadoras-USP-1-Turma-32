```python
tecla= input()
if tecla== 'a' or tecla== 'A':
  print('LEFT')
elif tecla== 'w' or tecla== 'W':
  print('UP')
elif tecla== 'd' or tecla== 'D':
  print('RIGHT')
elif tecla== 's' or tecla== 'S':
  print('DOWN')
elif tecla != 'a' or 'A' or 'w' or 'W' or 'd' or 'D' or 'S' or 's':
  print('?')
```
