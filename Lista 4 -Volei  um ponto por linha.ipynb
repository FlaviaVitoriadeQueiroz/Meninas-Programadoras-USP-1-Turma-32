```python
time1 = 0
time2 = 0
continuar = 1
while continuar == 1:
    ponto = int(input())
    if ponto == 1:
        time1 += 1
    elif ponto == 2:
        time2 += 1
    if (time1 >= 25 or time2 >= 25):
        if time1 > time2 + 1 or time2 > time1 + 1:
            continuar = 0
print(time1, "x", time2)
```
