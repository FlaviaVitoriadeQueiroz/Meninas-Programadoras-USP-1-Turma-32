```python
contador = 0
entrada = input()

while entrada != "CDA 1942":
    if entrada != "":
        contador += 1
    entrada = input()

print(contador)
```

```python
palavra = ''
entrada = input()
while entrada != "CDA":
    if entrada != "":
     entrada = input()
     palavra = entrada + palavra
print(palavra, '1942')
```
